1) Faça uma função que dado um vetor de inteiros retorne o maior inteiro do vetor.

```python
def MaiorValor(vetor):
  if (len(vetor) == 0):
    return None

  maior = vetor[0]

  for i in range(1, len(vetor)):
    if vetor[i] > maior:
      maior = vetor[i]

  return maior

vetor = [1, 2, 29, 7]

print(MaiorValor(vetor))
```

2) Você está organizando uma lista de alunos por turmas em uma escola. Cada turma contém vários alunos, e cada aluno é deve conter um nome, idade e id.

Exercício da aula passada, porém deve ser resolvido utilizando o que aprendemos sobre POO.

a) Crie duas turmas, cada uma contendo dois alunos.

b) Adicionar um novo aluno: Adicione um novo aluno em uma das turmas.

c) Remover um aluno: Remova um aluno de uma das turmas.

d) Exibir os alunos de uma turma

```python
class Aluno():
  def __init__(self, nome, idade, id):
    self.nome = nome
    self.idade = idade
    self.id = id
```

```python
class Turma():
  def __init__(self, nome):
    self.nome = nome
    self.alunos = []

  def AdicionarAluno(self, aluno):
    self.alunos.append(aluno)

  def RemoverAluno(self, idAlunoRemover):
    for aluno in self.alunos:
      if aluno.id == idAlunoRemover:
        self.alunos.remove(aluno)

  def ListarAlunos(self):
    if len(self.alunos) == 0:
      print("Essa turma não contem alunos")

      return None


    for aluno in self.alunos:
      print(f"Nome: {aluno.nome}. Idade: {aluno.idade}. Id: {aluno.id}")
```

```python
#a)

turmaA = Turma("turma a")
turmaB = Turma("turma b")

aluno1 = Aluno(nome="aluno1", idade=20, id=1)
aluno2 = Aluno(nome="aluno2", idade=22, id=2)

turmaA.AdicionarAluno(aluno1)
turmaA.AdicionarAluno(aluno2)

turmaB.AdicionarAluno(Aluno(nome="aluno3", idade=21, id=3))
turmaB.AdicionarAluno(Aluno(nome="aluno4", idade=28, id=4))

turmaA.ListarAlunos()

print()

turmaB.ListarAlunos()
```

```python
#b)

turmaA.AdicionarAluno(Aluno("aluno5", idade=30, id=5))
```

```python
#c)
turmaA.ListarAlunos()

turmaA.RemoverAluno(idAlunoRemover=2)

print()

turmaA.ListarAlunos()
```

```python
#d)

print(f"Alunos da turma: {turmaA.nome}")
turmaA.ListarAlunos()

print()

print(f"Alunos da turma: {turmaB.nome}")
turmaB.ListarAlunos()
```

3) Crei uma função que dado dois vetores retorne um novo vetor que é a concatenação de ambos os vetores.

```python
def Concatenar(vetor1, vetor2):
  NovoVetor = vetor1.copy()

  for elemento in vetor2:
    NovoVetor.append(elemento)

  return NovoVetor

vetor1 = [1, 2, 3]
vetor2 = [4, 5, 6]

NovoVetor = Concatenar(vetor1, vetor2)

print(NovoVetor)
print(vetor1)
```

4) Crie uma função que dado dois vetores retorne um novo vetor que é composto dos maiores elementos de cada vetor

```python
def Maiores(vetor1, vetor2):
  NovoVetor = []

  NovoVetor.append(MaiorValor(vetor1))
  NovoVetor.append(MaiorValor(vetor2))

  return NovoVetor

vetor1 = [1, 2, 3]
vetor2 = [100, 950, 2]

print(Maiores(vetor1, vetor2))
```

5) Crie uma função que receba uma lista de palavras e retorne o número de palavras que tenham um tamanho maior do que 5

```python
def PalavrasMaiores5(ListaPalavras):
  contador = 0

  for palavra in ListaPalavras:
    if len(palavra) > 5:
      contador += 1

  return contador

Palavras = ["Ola", "Paralelepipedo", "Python", "ICMC", "DATA", "DATAICMC"]

print(PalavrasMaiores5(Palavras))
```
